```python
import pandas as pd
```

```python
steamdf = pd.read_json("steamdb.json")
```

```python
## Change option to be able to view whole df (deciding columns desired)
pd.set_option('display.max_columns', None)
```

```python
## Deciding which columns to continue using
steamdf.head()

## sid (game id), store_uscore (score in percent), published_store (date published on steam), name,
## description, full_price, current_price, platforms, developers, publishers, language, categories,
## genres, tags, gfq_difficulty (gameFAQs difficulty), gfq_rating, gfq_length, meta_score (official), 
## meta_uscore (user score), igdb_score, igdb_uscore, igdb_popularity
```

```python
## new df with only the desired columns
steamdf = (pd.read_json("steamdb.json")
    .loc[:,["sid", "name", "published_store", "full_price", "current_price", "developers","publishers",
        "platforms","languages","categories","genres","tags","store_uscore","gfq_difficulty",
        "gfq_rating","gfq_length","meta_score","meta_uscore","igdb_score","igdb_uscore","igdb_popularity"]]
    .rename(columns={"published_store":"publish_date"})
)
steamdf.head()
```

```python
## look for opportunities to optimize data types
steamdf.info(verbose=True)

## publish_date to date
## full_price and current_price * 0.01 to convert to USD format
## convert platforms to category, 7
## convert gfq_difficulty to category, 9

steamdf["genres"].value_counts()
steamdf["gfq_difficulty"].nunique()
```

```python
## publish_date to date
steamdf["publish_date"] = pd.to_datetime(steamdf["publish_date"])

steamdf.info()
```

```python
## full_price and current_price * 0.01 to convert to USD format
steamdf["full_price"] = steamdf["full_price"] * 0.01
steamdf["current_price"] = steamdf["current_price"] * 0.01
```

```python
## convert platforms to category, 7
## convert gfq_difficulty to category, 9
steamdf["platforms"] = steamdf["platforms"].astype("category")
steamdf["gfq_difficulty"] = steamdf["gfq_difficulty"].astype("category")
steamdf.info()
```
